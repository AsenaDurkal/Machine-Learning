```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import joblib
import numpy as np
```

```python
#loading the dataset csv
X = pd.read_csv('X.csv')
```

```python
#Storing target variable
y = pd.read_csv('y.csv')
```

```python
#Split data into training and testing sets 
# 20% of the dataset will be used for test and the rest, 80% will be used for training
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)

y_train = np.ravel(y_train)
y_test = np.ravel(y_test)
print("Size of X_train", X_train.shape)
print("Size of X_test", X_test.shape)
print("Size of y_train", y_train.shape)
print("Size of y_test", y_test.shape)
```

```python
#Train and testing the different classifiers (80% training and 20% testing) 
# Had to increase the max iterations as i encountered an error: ConvergenceWarning:lbfgs failed to converge (status=1):STOP: TOTAL NO. of ITERATIONS REACHED Limit.

cls1=  LogisticRegression(solver = 'lbfgs', max_iter=10000)
cls2 =  RandomForestClassifier()
```

```python
cls1.fit(X_train, y_train)
cls2.fit(X_train, y_train)
```

```python
#predict
predictions1 = cls1.predict(X_test)
probabilities1 = cls1.predict_proba(X_test)

predictions2 = cls2.predict(X_test)
probabilities2 = cls2.predict_proba(X_test)
```

```python
# Calculating metrics with the calculated predictions and probability and y_test as a ground truth.
print(f"Metrics for  Logistic Regression:")
print(f"Accuracy: {accuracy_score(y_test, predictions1)}") #Number of correct predictions / Total Number of Oredictions
print(f"Precision: {precision_score(y_test, predictions1)}") # Correctnes of positive predictions made; True Positives/(True Positives + False Positives)
print(f"Recall: {recall_score(y_test, predictions1)}")    # Measuress the ability of model to catch positive instances; True Pisitive/ (True Positives + False Negatives)
print(f"F1 Score: {f1_score(y_test, predictions1)}")  # Harmonic mean of precision and recall (Precision * Recall) / (Precision+Recall)
print("\n")

print(f"Metrics for  Random Forest:")
print(f"Accuracy: {accuracy_score(y_test, predictions2)}") #Number of correct predictions / Total Number of Oredictions
print(f"Precision: {precision_score(y_test, predictions2)}") # Correctnes of positive predictions made; True Positives/(True Positives + False Positives)
print(f"Recall: {recall_score(y_test, predictions2)}")    # Measuress the ability of model to catch positive instances; True Pisitive/ (True Positives + False Negatives)
print(f"F1 Score: {f1_score(y_test, predictions2)}")  # Harmonic mean of precision and recall (Precision * Recall) / (Precision+Recall)
print("\n")



```

```python
# Exporting the trained models to different files with their name.
joblib.dump(cls1,"LogisticRegression_model"+".joblib")
joblib.dump(cls2,"RandomForest_model"+".joblib")
```
